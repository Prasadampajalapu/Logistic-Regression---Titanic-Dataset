## Logistic Regression - Titanic Dataset

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
train_data=pd.read_csv('training_titanic_x_y_train.csv')
test_data=pd.read_csv('test_titanic_x_test.csv')

print(train_data.shape)
print(test_data.shape)
```

```python
train_data.drop(['Name','Sex','Ticket','Cabin','Embarked'],axis=1,inplace=True)
test_data.drop(['Name','Sex','Ticket','Cabin','Embarked'],axis=1,inplace=True)
```

```python
train_data.isnull().sum()
```

```python
test_data.isnull().sum()
```

```python
train_data['Age'].fillna(method='pad',inplace=True)
```

```python
test_data['Age'].fillna(method='pad',inplace=True)
```

```python
train_data.isnull().sum()
```

```python
test_data.isnull().sum()
```

```python
x_train_data=train_data.iloc[:,:-1].values
y_train_data=train_data.iloc[:,-1].values
```

```python
model=LogisticRegression()
model.fit(x_train_data,y_train_data)
```

```python
y_pred=model.predict(test_data)
```

```python
#create csv file for the predictions
np.savetxt('predictions.csv',y_pred,fmt='%.1d',delimiter=',')
```
